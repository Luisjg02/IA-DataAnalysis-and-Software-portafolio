# **Algoritmo minimax + Alpha-beta pruning**

Implementación del algoritmo Minimax con poda Alpha-Beta.

Descripción:
Este programa evalúa un árbol de búsqueda adversarial donde:
- La raíz es un nodo MAX (quiere maximizar el resultado).
- Sus hijos son nodos MIN (quieren minimizar el resultado).
- Las hojas contienen valores numéricos finales.

Parámetros principales:
- alpha: mejor valor encontrado hasta el momento para MAX.
- beta: mejor valor encontrado hasta el momento para MIN.

La poda Alpha-Beta ocurre cuando beta <= alpha,
lo que permite evitar evaluar ramas que no afectarán
la decisión final.

Salida:
- Valor óptimo en la raíz.
- Lista de nodos que fueron podados.


Utilice el primer arbol


```python
tree = {
    "A1": [13, 11, 19, 11, 10],
    "A2": [15, 12, 10, 2, 12]
}

podados = []

def minimax(node, alpha, beta, maximizing):

    global podados

    # Nodo MAX
    if maximizing:
        value = float("-inf")
        for key in node:
            value = max(value, minimax(node[key], alpha, beta, False))
            alpha = max(alpha, value)

            if beta <= alpha:
                break  # poda

        return value

    # Nodo MIN
    else:
        value = float("inf")
        for i, v in enumerate(node):
            value = min(value, v)
            beta = min(beta, value)

            if beta <= alpha:
                podados.extend(node[i+1:])
                break

        return value


resultado = minimax(tree, float("-inf"), float("inf"), True)

print("Valor óptimo en la raíz:", resultado)
print("Nodos podados:", podados)
```

```python
"""
Juego: elegir el mejor cofre.
MAX elige el cofre.
MIN elige el peor premio dentro del cofre.
Se usa Minimax con poda Alpha-Beta.
"""

game = {
    "Cofre A": [8, 6, 7],
    "Cofre B": [5, 9, 4]
}

podados = []

def minimax(node, alpha, beta, maximizing):

    global podados

    # Nodo MAX
    if maximizing:
        mejor_valor = float("-inf")
        mejor_jugada = None

        for cofre in node:
            valor = minimax(node[cofre], alpha, beta, False)

            if valor > mejor_valor:
                mejor_valor = valor
                mejor_jugada = cofre

            alpha = max(alpha, mejor_valor)

            if beta <= alpha:
                break

        return mejor_valor, mejor_jugada

    # Nodo MIN
    else:
        peor_valor = float("inf")

        for i, premio in enumerate(node):
            peor_valor = min(peor_valor, premio)
            beta = min(beta, peor_valor)

            if beta <= alpha:
                podados.extend(node[i+1:])
                break

        return peor_valor


valor, jugada = minimax(game, float("-inf"), float("inf"), True)

print("Mejor cofre:", jugada)
print("Valor asegurado:", valor)
print("Premios podados:", podados)
```
