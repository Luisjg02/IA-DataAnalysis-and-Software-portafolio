```python
#Luis Jose Gonzalez Montilla
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Crear el Dataset
data = np.array([
    [1, 1],
    [1.5, 2],
    [3, 4],
    [5, 7],
    [3.5, 5],
    [4.5, 5]
])

df = pd.DataFrame(data, columns=["X", "Y"])

kmeans_model = KMeans(n_clusters=2, random_state=42, n_init=10)

df["Cluster"] = kmeans_model.fit_predict(data)

centroids = kmeans_model.cluster_centers_

print("Data with assigned clusters:")
print(df)

print("\nCentroids:")
print(centroids)

# clusters
plt.figure(figsize=(8, 6))
plt.scatter(df["X"], df["Y"], c=df["Cluster"], s=100, label="Data points")
plt.scatter(centroids[:, 0], centroids[:, 1], marker="X", s=200, label="Centroids")

plt.title("K-means Clustering Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()
plt.show()
```
