```python
#Luis Jose Gonzalez Montilla
# Generalidades de Machine Learning

# 1. Importar librerías
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2. Cargar el conjunto de datos
data = load_iris()
X = data.data      # Características
y = data.target    # Etiquetas

# 3. División de datos
# Entrenamiento: 70%
# Prueba: 30%
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4. Crear el modelo (algoritmo)
model = LogisticRegression(max_iter=200)

# 5. Entrenamiento del modelo
model.fit(X_train, y_train)

# 6. Predicciones
y_pred = model.predict(X_test)

# 7. Evaluación del modelo (métricas)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
```

```python
# ================================
# Machine Learning con Árbol de Decisión
# ================================

# 1. Importar librerías
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2. Cargar datos
data = load_iris()
X = data.data
y = data.target

# 3. Dividir datos (entrenamiento y prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 4. Crear el modelo Árbol de Decisión
model = DecisionTreeClassifier(random_state=42)

# 5. Entrenar el modelo
model.fit(X_train, y_train)

# 6. Realizar predicciones
y_pred = model.predict(X_test)

# 7. Evaluación del modelo
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

data = load_iris()
X, y = data.data, data.target

# Entrenamiento
X_train, _, y_train, _ = train_test_split(
    X, y, test_size=0.5, random_state=1
)

# Testeo con otro conjunto diferente
_, X_test, _, y_test = train_test_split(
    X, y, test_size=0.5, random_state=99
)

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# ================================
# Dataset Wine - Machine Learning
# ================================

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Cargar dataset
data = load_wine()
X = data.data
y = data.target

# 2. División entrenamiento / prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3. Crear modelo
model = DecisionTreeClassifier(random_state=42)

# 4. Entrenar
model.fit(X_train, y_train)

# 5. Predecir
y_pred = model.predict(X_test)

# 6. Evaluar
print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred))
```

```python
# ======================================
# 1. COMPONENTES DEL MODELO
# ======================================

# Datos
from sklearn.datasets import load_wine
data = load_wine()
X = data.data
y = data.target

# Algoritmo
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)

# Métricas
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ======================================
# 2. FLUJO DE TRABAJO
# ======================================

# División de datos (entrenamiento y prueba)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Entrenamiento
model.fit(X_train, y_train)

# Prueba
y_pred = model.predict(X_test)

# Evaluación
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))
print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred))

# ======================================
# 3. PROBLEMAS COMUNES (ANÁLISIS)
# ======================================

# Comparación de rendimiento para detectar sobreajuste o subajuste
train_accuracy = model.score(X_train, y_train)
test_accuracy = model.score(X_test, y_test)

print("\nAccuracy entrenamiento:", train_accuracy)
print("Accuracy prueba:", test_accuracy)
```
