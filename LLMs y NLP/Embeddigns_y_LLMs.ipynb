```python
!pip install --upgrade datasets sentence-transformers scikit-learn matplotlib seaborn pandas numpy

import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, confusion_matrix

try:
    from sklearn.metrics import root_mean_squared_error as rmse_func
    HAS_RMSE = True
except ImportError:
    HAS_RMSE = False
    def rmse_func(y_true, y_pred):
        return np.sqrt(mean_squared_error(y_true, y_pred))
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import KBinsDiscretizer

# 1. Recolección de la data
"""
Carga dataset validado: smartcat/Amazon_Beauty_and_Personal_Care_2023 (belleza, Parquet nativo, sin scripts).
Columnas: 'description' (texto como reseñas), 'price' (float USD), 'brand'/'categories' (features).
Filtra >2000 válidos con precio >0. Fuente: https://huggingface.co/datasets/smartcat/Amazon_Beauty_and_Personal_Care_2023
"""
dataset = load_dataset("smartcat/Amazon_Beauty_and_Personal_Care_2023", split="train")
df = pd.DataFrame(dataset)

# Ajusta columnas reales
df = df.rename(columns={'description': 'text'})
df['text'] = df['text'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x))
df = df.dropna(subset=['text', 'price'])
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])
df = df[df['price'] > 0]
df = df.head(10000)  # Submuestra para pruebas (>2000); quita para full
print(f"Dataset: {len(df)} filas. Ejemplo precio: {df['price'].iloc[0]}")

# 2. Preparación / preprocesamiento
"""
Limpieza: Str texto, float precio. Balanceo deciles para prevalencia (equilibrado, bajo sesgo).
Proporción: 80/20 train/test.
"""
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Precios Antes Balanceo')
plt.show()

df['price_decile'] = pd.qcut(df['price'], 10, labels=False, duplicates='drop')
min_count = df['price_decile'].value_counts().min()
balanced_df = df.groupby('price_decile').apply(lambda x: x.sample(min(min_count, len(x)), random_state=42)).reset_index(drop=True)
print(f"Balanceado: {len(balanced_df)} filas.")

plt.figure(figsize=(10, 5))
sns.histplot(balanced_df['price'], bins=50, kde=True)
plt.title('Precios Después Balanceo')
plt.show()

test_size = 0.2

# 3 LLMs para embeddings
"""
3 LLMs: all-MiniLM-L6-v2, all-mpnet-base-v2, paraphrase-mpnet-base-v2.
"""
embedder_names = ['all-MiniLM-L6-v2', 'all-mpnet-base-v2', 'paraphrase-mpnet-base-v2']
embedders = [SentenceTransformer(name) for name in embedder_names]

# Modelos de regresión
"""
Descripciones/Hiperparámetros: Defaults.
- KNN: n_neighbors=5
- RF: n_estimators=100
- DT: max_depth=None
- LS: OLS
- SVR: kernel='rbf'
- SGD: max_iter=1000
- AdaBoost: n_estimators=50
"""
models = {
    'KNN': KNeighborsRegressor(n_neighbors=5),
    'RF': RandomForestRegressor(n_estimators=100, random_state=42),
    'DT': DecisionTreeRegressor(random_state=42),
    'LS': LinearRegression(),
    'SVR': SVR(kernel='rbf'),
    'SGD': SGDRegressor(random_state=42, max_iter=1000),
    'AdaBoost': AdaBoostRegressor(n_estimators=50, random_state=42)
}

# 3. Entrenamiento / 4. Validación (21 pruebas)
results = []
for embedder_idx, embedder in enumerate(embedders):
    print(f"Embedder: {embedder_names[embedder_idx]}")
    embeddings = embedder.encode(balanced_df['text'].tolist(), show_progress_bar=True, batch_size=32, normalize_embeddings=True)
    X, y = embeddings, balanced_df['price'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    for model_name, model in models.items():
        start_time = time.time()
        model.fit(X_train, y_train)
        train_time = time.time() - start_time
        start_time = time.time()
        y_pred = model.predict(X_test)
        inf_time = (time.time() - start_time) / len(X_test)

        mae = mean_absolute_error(y_test, y_pred)
        rmse = rmse_func(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Params approx
        if hasattr(model, 'coef_'):
            params = model.coef_.size + getattr(model, 'intercept_', np.array([0])).size
        elif hasattr(model, 'estimators_'):
            if model_name == 'RF':
                params = len(model.estimators_) * model.estimators_[0].tree_.node_count
            else:
                params = sum(est.tree_.node_count for est in model.estimators_)
        else:
            params = X.shape[1] if model_name == 'KNN' else 'N/A'
        model_size_mb = embeddings.nbytes / (1024**2)

        results.append({
            'Embedder': embedder_names[embedder_idx],
            'Model': model_name,
            'MAE': mae,
            'RMSE': rmse,
            'R2': r2,
            'Train Time (s)': round(train_time, 2),
            'Inf Time (s/sample)': round(inf_time, 4),
            'Params': params,
            'Size (MB)': round(model_size_mb, 2)
        })

        # Matriz confusión (deciles)
        binner = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='quantile')
        y_test_bins = binner.fit_transform(y_test.reshape(-1, 1)).ravel()
        y_pred_bins = binner.transform(y_pred.reshape(-1, 1)).ravel()
        cm = confusion_matrix(y_test_bins, y_pred_bins)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'Confusión - {model_name} + {embedder_names[embedder_idx]}')
        plt.show()

        print(f"{model_name}: MAE={mae:.2f}, RMSE={rmse:.2f}, R2={r2:.2f}")

# Benchmark
benchmark_df = pd.DataFrame(results)
print(benchmark_df)
benchmark_df.to_csv('benchmark_results.csv', index=False)

# 5. Despliegue
best_result = benchmark_df.loc[benchmark_df['R2'].idxmax()]
print(f"Mejor: {best_result['Model']} con {best_result['Embedder']} (R2={best_result['R2']:.3f})")
best_embedder = SentenceTransformer(best_result['Embedder'])
best_model = models[best_result['Model']]
embeddings_full = best_embedder.encode(balanced_df['text'].tolist(), normalize_embeddings=True)
X_train_full, _, y_train_full, _ = train_test_split(embeddings_full, balanced_df['price'], test_size=test_size, random_state=42)
best_model.fit(X_train_full, y_train_full)

# Nueva data
new_text = ["Luxurious cream that hydrates deeply, perfect for dry skin."]
new_embedding = best_embedder.encode([str(new_text[0])], normalize_embeddings=True)
pred_price = best_model.predict(new_embedding)
print(f"Predicho: ${pred_price[0]:.2f}")

"""
Sistema: Balanceado/limpiado -> train (80%, embeddings -> precio) / test (20%, métricas/confusión).
Nueva data: Limpia/encode/predict. Prevalencia: Deciles equilibran.
"""
```
