```python
# REGRESIÓN LINEAL MÚLTIPLE
# Luis Jose Gonzalez Montilla

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
import seaborn as sns

# Función para generar y analizar datos
def generar_datos_y_analizar(nombre, y_var, x_vars, generador, n, random_state):
    # Generar datos
    df = generador(n=n, random_state=random_state)

    # Ajustar modelo
    formula = f"{y_var} ~ {' + '.join(x_vars)}"
    modelo = smf.ols(formula, data=df).fit()

    # Calcular matriz de correlación
    correlacion = df[[y_var] + x_vars].corr()

    # Guardar imagen de correlación
    plt.figure(figsize=(6, 4))
    sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(f"Matriz de Correlación - {nombre}")
    plt.tight_layout()
    img_path = f"correlacion_{nombre}.png"
    plt.savefig(img_path)
    plt.close()

    # ANOVA
    anova_res = anova_lm(modelo)

    # Imprimir resultados
    print(f"\n=== {nombre} ===")
    print("Vista previa de datos:")
    print(df.head(), "\n")
    print("Resumen del modelo:")
    print(modelo.summary(), "\n")
    print("ANOVA:")
    print(anova_res, "\n")
    print(f"Imagen de correlación guardada en: {img_path}")

    return df, modelo, correlacion, anova_res, img_path

# Generadores de datos

def generar_ej1(n=50, random_state=42):
    np.random.seed(random_state)
    horas = np.random.normal(10, 3, n).clip(0)
    padres = np.random.randint(1, 6, n)
    ingreso = np.random.normal(1500, 500, n).clip(500)
    rendimiento = (50 + 2*horas + 3*padres + 0.005*ingreso +
                   np.random.normal(0, 5, n)).clip(0, 100)
    return pd.DataFrame({
        "Rendimiento": rendimiento,
        "Horas_estudio": horas,
        "Nivel_padres": padres,
        "Ingreso_familiar": ingreso
    })

def generar_ej2(n=50, random_state=43):
    np.random.seed(random_state)
    salud = np.random.normal(8, 2, n).clip(1)
    contaminacion = np.random.normal(25, 5, n).clip(5)
    empleo = np.random.normal(70, 10, n).clip(30, 100)
    calidad = (40 + 1.5*salud - 0.8*contaminacion + 0.3*empleo +
               np.random.normal(0, 4, n)).clip(0, 100)
    return pd.DataFrame({
        "Calidad_vida": calidad,
        "Acceso_salud": salud,
        "Contaminacion": contaminacion,
        "Empleo": empleo
    })

def generar_ej3(n=100, random_state=44):
    np.random.seed(random_state)
    ingreso = np.random.normal(2000, 600, n).clip(500)
    historial = np.random.normal(70, 15, n).clip(0, 100)
    edad = np.random.normal(35, 10, n).clip(18, 70)
    monto = (500 + 0.8*ingreso + 50*historial - 10*edad +
             np.random.normal(0, 1000, n)).clip(0)
    return pd.DataFrame({
        "Monto_prestamo": monto,
        "Ingreso": ingreso,
        "Historial": historial,
        "Edad": edad
    })

def generar_ej4(n=50, random_state=45):
    np.random.seed(random_state)
    imc = np.random.normal(26, 4, n).clip(15, 40)
    edad = np.random.normal(50, 15, n).clip(18, 90)
    sal = np.random.normal(8, 3, n).clip(0, 20)
    presion = (90 + 0.9*imc + 0.4*edad + 0.8*sal +
               np.random.normal(0, 5, n)).clip(70, 200)
    return pd.DataFrame({
        "Presion": presion,
        "IMC": imc,
        "Edad": edad,
        "Sal": sal
    })

# Ejecutar análisis para los 4 ejercicios
df1, modelo1, cor1, anova1, img1 = generar_datos_y_analizar(
    "Ejercicio1", "Rendimiento", ["Horas_estudio", "Nivel_padres", "Ingreso_familiar"], generar_ej1, 50, 42)

df2, modelo2, cor2, anova2, img2 = generar_datos_y_analizar(
    "Ejercicio2", "Calidad_vida", ["Acceso_salud", "Contaminacion", "Empleo"], generar_ej2, 50, 43)

df3, modelo3, cor3, anova3, img3 = generar_datos_y_analizar(
    "Ejercicio3", "Monto_prestamo", ["Ingreso", "Historial", "Edad"], generar_ej3, 100, 44)

df4, modelo4, cor4, anova4, img4 = generar_datos_y_analizar(
    "Ejercicio4", "Presion", ["IMC", "Edad", "Sal"], generar_ej4, 50, 45)

print("\nTodas las imágenes han sido guardadas.")

```
