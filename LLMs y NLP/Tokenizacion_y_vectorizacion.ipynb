```python
# Tokenizacion y vectorizaccion
# 1 Datos de entrada
comentarios = [
    "La mejor pizza que he comido en años. Literalmente sin desperdicios, excelente sabor y textura. 100% recomendado.",
    "Siento que la salsa de tomate estaba muy dulce. Terrible combinación con queso de cabra, habría preferido queso mozzarella normal.",
    "Menos es más. Fue tan terrible que solo me comí la punta de un pedazo. Demasiados ingredientes, es coo si fuera para engordar animales.",
    "La idea está bien, pero la ejecución no fue la mejor. Lo que hace que esta pizza no termine de ser terrible son los embutidos. La alta de calidad del pepperoni para mí la salvó, lo demás, a la basura.",
    "Me encantan las pizzas como estas. Esta incorporación de sabores exóticos fue increíble. De verdad es una experiencia que todos deben degustar.",
    "Excelente pizza, nunca había probado dátiles en una pizza y me sorprendió que combinara tan bien con el queso de cabra. Será una de las mejores pizzas en mi top.",
    "La compré con miedo porque había leído muchas reviews malas, pero al probarla quedé enamorado. ¡Qué delicia de pizza, cuánto sabor!",
    "Horrible el queso de cabra y no pega para nada con los dátiles. Siento que quisieron hacerla muy compleja y al final terminó siendo un desastre.",
    "La pizza definitivamente es deliciosa, una de las mejores que he probado. La textura de la arena fue delicada y crujiente, lo único es que siento que son muy caras para el producto que ofrecen.",
    "Me gustó mucho el sabor. Ya quiero que sea viernes otra vez para volver.",
    "¡Qué deliciosa pizza! Si no estuviera a dieta me verían todos los días aquí.",
    "Esta pizza es una basura, un desastre total. No la recomiendo, mejor vayan a otro sitio. Para colmo carísimas."
]

sentimientos = [
    "Positivo","Negativo","Negativo","Negativo","Positivo","Positivo",
    "Positivo","Negativo","Positivo","Positivo","Positivo","Negativo"
]

 # 2 Stopwords personalizadas
stop_words = {
    'la','el','los','las','un','una','unos','unas','que','de','del','al','a','en','y','o','por','para','se','es',
    'con','su','sus','lo','le','les','me','mi','mis','te','tu','tus','ya','como','pero','muy','no','si','más','fue',
    'esta','este','estas','estos','son','ha','he','han','hace','porque','qué','quien','cual','cuales','cuando','donde',
    'todo','todos','toda','todas','esa','ese','esas','esos','ser','soy','eres','era','eran','está','están','estaba'
}

# 3 Limpieza y tokenización
def limpiar_tokens(texto):
    texto = texto.lower()
    tokens = re.findall(r'\b[a-záéíóúüñ]+\b', texto)  # Solo palabras (sin números ni signos)
    tokens = [t for t in tokens if t not in stop_words]  # Eliminar stopwords
    return tokens

comentarios_tokens = [limpiar_tokens(c) for c in comentarios]

# 4 Crear vocabulario
vocabulario = sorted(set([t for tokens in comentarios_tokens for t in tokens]))

# 5️ Binary Term Frequency
binary_data = []
for tokens in comentarios_tokens:
    fila = [1 if palabra in tokens else 0 for palabra in vocabulario]
    binary_data.append(fila)
df_binary = pd.DataFrame(binary_data, columns=vocabulario)

# 6️ Bag of Words
bow_data = []
for tokens in comentarios_tokens:
    fila = [tokens.count(palabra) for palabra in vocabulario]
    bow_data.append(fila)
df_bow = pd.DataFrame(bow_data, columns=vocabulario)

# 7️ TF-IDF manual
N = len(comentarios_tokens)
idf = {}
for palabra in vocabulario:
    df = sum(1 for tokens in comentarios_tokens if palabra in tokens)
    idf[palabra] = log(N / (1 + df)) + 1  # Evita división por cero

tfidf_data = []
for tokens in comentarios_tokens:
    fila = []
    for palabra in vocabulario:
        tf = tokens.count(palabra)
        fila.append(tf * idf[palabra])
    tfidf_data.append(fila)
df_tfidf = pd.DataFrame(tfidf_data, columns=vocabulario)

# 8️ Unir todo en un solo DataFrame
df_final = pd.concat(
    [pd.DataFrame({'Comentario': comentarios, 'Sentimiento': sentimientos}),
     df_binary.add_prefix('Binary_'),
     df_bow.add_prefix('BoW_'),
     df_tfidf.add_prefix('TFIDF_')],
    axis=1
    )
```
