```python
# Leer la hoja específica
df = pd.read_excel("Nomina.xlsx", sheet_name="NOMINA PORTAL FIJO ABRIL 2025", skiprows=12)

df.columns = df.iloc[0]  # toma esa fila y la usa como nombres de columnas
df = df[1:]              # luego elimina esa fila de los datos reales

df.head(40)

# Rellenar los valores nulos de una sola columna con 0
df['SFS SALUD ADICIONAL'] = df['SFS SALUD ADICIONAL'].fillna(0)
df.head(40)

df.info()

# Eliminar filas que están completamente vacías
df = df.dropna(how='all')

# O eliminar filas que tienen muchos NaN (por ejemplo, más de la mitad vacías)
df = df.dropna(thresh=len(df.columns)//2)

# Rellenar solo columnas numéricas con cero
columnas_numericas = ['SALARIO', 'SEGURO DE VIDA', 'IMPUESTO SOBRE LA RENTA', 'AFP', 'SEGURO FAMILIAR DE SALUD SFS', 'TOTAL DESCUENTOS', 'SUELDO NETO']
df[columnas_numericas] = df[columnas_numericas].fillna(0)

# Quitar filas donde la columna 'NO' (o la que tenga los números) diga TOTAL o esté vacía
df = df[~df['NO'].astype(str).str.upper().str.contains("TOTAL", na=False)]

# También puedes eliminar filas donde el nombre está vacío (opcional)
df = df[df['NOMBRE Y APELLIDO'].notna()]

# Rellenar con 0 en las columnas numéricas
columnas_numericas = ['SALARIO', 'SEGURO DE VIDA', 'IMPUESTO SOBRE LA RENTA', 'AFP', 'SEGURO FAMILIAR DE SALUD SFS', 'TOTAL DESCUENTOS', 'SUELDO NETO']
df[columnas_numericas] = df[columnas_numericas].fillna(0)

# Rellenar texto con "Sin dato"
df = df.fillna("Sin dato")

df.tail(10)

df.isna().sum()
```
