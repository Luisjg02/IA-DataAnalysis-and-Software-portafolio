# LIbrerias

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

# 1. Imagen en blanco con fondo negro

```python
def create_image(size=(500, 500)):
    return np.zeros((*size, 3), dtype=np.uint8)

image = create_image()
cv2_imshow(image)
```

#2. Dibujar Figuras Geométricas

# Linea azul

```python
def draw_shapes(image):
    cv2.line(image, (0, 0), (499, 499), (255, 0, 0), 3)

draw_shapes(image)
cv2_imshow(image)
```

# Rectangulo verde

```python
def draw_shapes(image):
    cv2.rectangle(image, (150, 150), (350, 350), (0, 255, 0), 3)

draw_shapes(image)
cv2_imshow(image)
```

# Circulo rojo

```python
def draw_shapes(image):
    cv2.circle(image, (250, 250), 50, (0, 0, 255), -1)

draw_shapes(image)
cv2_imshow(image)
```

#3. Agregar Texto

```python
def add_text(image, text="Luis Gonzalez 2024-0509"):
    cv2.putText(image, text, (30, 400), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 150, 100), 2)

add_text(image)
cv2_imshow(image)
```

#4. Convertir a Escala de Grises

```python
def convert_to_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

gray_image = convert_to_grayscale(image)
cv2_imshow(gray_image)
```

#5. Redimensionar Imagen

```python
def resize_image(image, scale=2):
    return cv2.resize(image, (image.shape[1] * scale, image.shape[0] * scale))

resized_image = resize_image(image)
cv2_imshow(resized_image)
```

#6. Rotar Imagen 45 Grados

```python
def rotate_image(image, angle):
    h, w = image.shape[:2]
    matrix = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1.0)
    return cv2.warpAffine(image, matrix, (w, h))

rotated_image = rotate_image(image, 45)
cv2_imshow(rotated_image)
```

#7. Rotar Imagen 30 Grados

```python
def rotate_image(image, angle):
    h, w = image.shape[:2]
    matrix = cv2.getRotationMatrix2D((w // 2, h // 2), angle, 1.0)
    return cv2.warpAffine(image, matrix, (w, h))

rotated_image = rotate_image(image, 30)
cv2_imshow(rotated_image)
```

#8. Transformación de Perspectiva

```python
def apply_perspective_transform(image):
    h, w = image.shape[:2]
    src = np.float32([[0, 0], [w-1, 0], [0, h-1], [w-1, h-1]])
    dst = np.float32([[90, 100], [w-30, 20], [20, h-60], [w-20, h-20]])
    matrix = cv2.getPerspectiveTransform(src, dst)
    return cv2.warpPerspective(image, matrix, (w, h))

perspective_image = apply_perspective_transform(image)
cv2_imshow(perspective_image)
```

#9. Crear Máscara de Transparencia

```python
def create_transparency_mask(image):
    mask = np.zeros(image.shape[:2], dtype=np.uint8)
    cv2.circle(mask, (250, 250), 110, 255, -1)
    return cv2.bitwise_and(image, image, mask=mask)

masked_image = create_transparency_mask(image)
cv2_imshow(masked_image)
```
